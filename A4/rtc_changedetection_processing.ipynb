```python
from rtc_changedetection import (
    load_rtc_image,
    normalize_image,
    stack_rtc_images,
    crop_to_common_extent,
    colorize_rtc_stack,
    display_change
)
```

```python
# Define the file paths for the 3 images

img1_path = "./data/month1.tif"
img2_path = "./data/month2.tif"
img3_path = "./data/month3.tif"
```

```python
# Load and normalize
img1 = normalize_image(load_rtc_image(img1_path))
img2 = normalize_image(load_rtc_image(img2_path))
img3 = normalize_image(load_rtc_image(img3_path))
```

```python
# Auto-crop them to the smallest common extent
img1_c, img2_c, img3_c = crop_to_common_extent(img1, img2, img3)
```

```python
# Stack and colorize
stack = stack_rtc_images(img1_c, img2_c, img3_c)
rgb = colorize_rtc_stack(stack)
```

```python
# Display

display_change(rgb)
```

**How to Intrepret the image**

**Reddish areas** : higher backscatter in month 1 

**Greenish areas** : higher backscatter in month 2

**Bluish areas** :higher backscatter in month 3

**White / Gray** : similar backscatter in all three months

**Black** : very low backscatter across all three months
